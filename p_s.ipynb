```python
#answer of Q.no.1
import math

#calculating the mean of the number
def mean(num):
  mean_value=sum(num)/len(num)
  return mean_value

# calculating meadian
def median(num):
  num.sort() #sorting the number for
  n = len(num)
  if n%2==0:
    median_value = (num[n//2]+ num[n//2 - 1])/2 #  even value
  else:
      median_value =(num[n//2]) # odd value
  return median_value

#calculating the standard derviation
def std_deviation(num):
    mean_s = mean(num)
    squared_diff = [(x - mean_s) ** 2 for x in num]
    variance = sum(squared_diff) / len(num)
    std_deviation_value = math.sqrt(variance)
    return std_deviation_value

#generating the random numbers for 500.
import random
random_num = [random.uniform(-20, 20) for _ in range(500)]

# Calculating the  mean, median, and standard deviation
mean_value = mean(random_num)
median_value =median(random_num)
std_deviation_value = std_deviation(random_num)

# printing the results
print("Mean:", mean_value)
print("Median:", median_value)
print("Standard Deviation:", std_deviation_value)
import matplotlib.pyplot as plt

# Creating  a histogram of random_num with 10 bins and black edges
plt.hist(random_num, bins=10, edgecolor='black')

# Label the x and y axes of the plot
plt.xlabel('Value')
plt.ylabel('Frequency')

# Set a title for the histogram plot
plt.title('Histogram of Random Numbers')

# Overlay vertical dashed lines for the mean, median, and standard deviation
plt.axvline(mean_value, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean_value:.2f}')
plt.axvline(median_value, color='green', linestyle='dashed', linewidth=2, label=f'Median: {median_value:.2f}')
plt.axvline(std_deviation_value, color='blue', linestyle='dashed', linewidth=2, label=f'Standard Deviation: {std_deviation_value:.2f}')

# Display a legend to identify the lines
plt.legend()

# Showing the plot
plt.show()

```

```python
#answer of Q.no.3
import matplotlib.pyplot as plt

# inputing the data
causes_of_death = [
    "Alzheimer's",
    "Chronic Respiratory Disease",
    "Diabetes",
    "Heart Disease",
    "Influenza/Pneumonia",
    "Malignant Neoplasma",
    "Accidents",
    "Nephritis/Nephrosis",
    "Septicemia",
    "Stroke"
]

num_of_deaths = [7.2,12.5,7.2,63.2,5.6,56.0,12.2,4.5,3.4,13.7]

# Calculate the cumulative percentage
total_deaths = sum(num_of_deaths)
cumulative_percentage = [100 * (sum(num_of_deaths[:i+1]) / total_deaths) for i in range(len(num_of_deaths))]

# Create a Pareto chart
fig, ax1 = plt.subplots(figsize=(22, 11))

# Bar chart
ax1.bar(causes_of_death, num_of_deaths, color='b')
ax1.set_xlabel('Causes of Death', fontsize=12)  # Adjust font size
ax1.set_ylabel('Number of Deaths (x 10,000)', color='b', fontsize=12)
ax1.tick_params('y', colors='b')

# Create a second y-axis for the cumulative percentage line
ax2 = ax1.twinx()
ax2.plot(causes_of_death, cumulative_percentage, marker='o', color='r', linestyle='-')
ax2.set_ylabel('Cumulative Percentage', color='r', fontsize=12)
ax2.tick_params('y', colors='r')

# Rotate x-axis labels for better readability
plt.xticks(rotation=90,fontsize=10, va="center")

# Title and labels
plt.title('Pareto Chart of Leading Causes of Death in the US (2006)')
plt.show()
```

```python
#answer of q.no.4
import numpy as np
from scipy import stats
from collections import Counter

data = [
    11, 14, 15, 15, 16, 16, 17, 18, 19, 21, 25, 36,
    12, 14, 15, 15, 16, 16, 17, 18, 19, 21, 25, 39,
    13, 14, 15, 15, 16, 17, 17, 18, 20, 22, 26, 43,
    13, 14, 15, 15, 16, 17, 17, 18, 20, 22, 26, 46,
    13, 14, 15, 16, 16, 17, 17, 18, 20, 22, 27, 50,
    13, 14, 15, 16, 16, 17, 17, 19, 20, 23, 27, 54,
    13, 14, 15, 16, 16, 17, 18, 19, 20, 23, 29, 59,
    13, 15, 15, 16, 16, 17, 18, 19, 20, 23, 30, 67,
    14, 15, 15, 16, 16, 17, 18, 19, 21, 24, 31,
    14, 15, 15, 16, 16, 17, 18, 19, 21, 24, 34
]

# Sort the data
sorted_data = sorted(data)

#calclutaing the mdeian
median = np.median(sorted_data)
print(f"Median: {median}")

# Count the occurrences of each element
counted_data = Counter(sorted_data)

# Find the modes (all modes)
modes = [k for k, v in counted_data.items() if v == max(counted_data.values())]

# Get the count of occurrences of the mode(s)
max_count = max(counted_data.values())

if len(modes) == 1:
    print(f"Mode: {modes[0]} (occurs {max_count} times)")
else:
    print(f"Modes: {', '.join(map(str, modes))} (each occurs {max_count} times)")

#For quartiles (Q1 and Q3):
q1 = np.percentile(sorted_data, 25)
q3 = np.percentile(sorted_data, 75)
print(f"Q1: {q1}")
print(f"Q3: {q3}")

# for percentiles (P10 and P95):
p10 = np.percentile(sorted_data, 10)
p95 = np.percentile(sorted_data, 95)
print(f"P10: {p10}")
print(f"P95: {p95}")
```
